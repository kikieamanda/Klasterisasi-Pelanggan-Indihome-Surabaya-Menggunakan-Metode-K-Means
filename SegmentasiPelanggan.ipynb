```python
import pandas as pd
```

```python
order = pd.read_csv("orderindihome.csv")
order
```

```python
order.info()
```

```python
order["TANGGAL ORDER"] = pd.to_datetime(order["TANGGAL ORDER"], format='%d/%m/%Y').dt.strftime('%m/%d/%Y')

order
```

```python
# Preprocessing: Convert TANGGAL ORDER to datetime and calculate duration since order
order['TANGGAL ORDER'] = pd.to_datetime(order['TANGGAL ORDER'])
order['LAMA LANGGANAN (BULAN)'] = (pd.Timestamp.now() - order['TANGGAL ORDER']).dt.days / 30
order
```

```python
# Menghapus baris yang mengandung "USULAN BRANCH" pada kolom "ID ORDER"
order = order[order["ID ORDER"] != "USULAN BRANCH"]
order
```

```python
order_cleaned = order.dropna()

order_cleaned
```

```python
# Membulatkan nilai pada kolom "LAMA LANGGANAN (BULAN)" menjadi bilangan bulat
order_cleaned["LAMA LANGGANAN (BULAN)"] = order_cleaned["LAMA LANGGANAN (BULAN)"].round().astype(int)

order_cleaned
```

```python
order_cleaned["HARGA PAKET PERBULAN"] = order_cleaned["HARGA PAKET PERBULAN"].str.replace(r"[^\d.]", "", regex=True).astype(float)
```

```python
order_cleaned["ID ORDER"] = order_cleaned["ID ORDER"].str.replace(r"[^\d.]", "", regex=True).astype(float)
```

```python
order_cleaned["LAMA LANGGANAN (BULAN)"] = order_cleaned["LAMA LANGGANAN (BULAN)"].astype(str).str.replace(r"[^\d.]", "", regex=True).astype(float)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
label_encoder = LabelEncoder()
order_cleaned['ENCODE PAKET'] = label_encoder.fit_transform(order_cleaned['PAKET INDIHOME'])
```

```python
features = order_cleaned[['HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)', 'ENCODE PAKET']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
order_cleaned['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=order_cleaned, x='HARGA PAKET PERBULAN', y='ENCODE PAKET', hue='ENCODE PAKET', palette='viridis')
plt.title('Scatter Plot antara Harga Paket Perbulan dan Encode Paket')
plt.xlabel('Harga Paket Perbulan')
plt.ylabel('Encode Paket')
plt.legend(title='Encode Paket')
plt.show()
```

```python
# Pair plot untuk semua fitur numerik
sns.pairplot(order_cleaned.drop(['ID ORDER'], axis=1), hue='PAKET INDIHOME')
plt.show()
```

```python
inertia = []
range_clusters = range(1, 11)

for k in range_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Menampilkan plot Elbow Method
plt.plot(range_clusters, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_scores.append(score)

# Plot silhouette score
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot metode elbow
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Metode Elbow WCSS')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Visualisasi hasil clustering menggunakan plot 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(data=order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', hue='Cluster', palette='viridis')
plt.title('KMeans Clustering')
plt.xlabel('Harga Paket Perbulan')
plt.ylabel('LAMA LANGGANAN (BULAN)')
plt.legend(title='Cluster')
plt.show()
```

```python
# Dari plot elbow dan silhouette score, kita pilih jumlah cluster optimal, misalnya 4
optimal_clusters = 3

# Implementasi KMeans dengan jumlah cluster optimal
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(scaled_features)
order_cleaned['Cluster'] = kmeans.labels_
```

```python
# Implementasi KMeans
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
order_cleaned['Cluster_KMeans'] = kmeans.fit_predict(scaled_features)

# Visualisasi interaktif dengan Plotly
fig = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_KMeans', title='KMeans Clustering', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig.show()
```

```python
# Implementasi DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
order_cleaned['Cluster_DBSCAN'] = dbscan.fit_predict(scaled_features)

# Implementasi Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=optimal_clusters)
order_cleaned['Cluster_Hierarchical'] = hierarchical.fit_predict(scaled_features)
```

```python
# Visualisasi DBSCAN dengan Plotly
fig_dbscan = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_DBSCAN', title='DBSCAN Clustering', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig_dbscan.show()

# Visualisasi Hierarchical Clustering dengan Plotly
fig_hierarchical = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_Hierarchical', title='Hierarchical Clustering', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig_hierarchical.show()
```

```python
# Preprocessing pipeline
preprocessing_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

# Clustering pipeline dengan KMeans
clustering_pipeline_kmeans = Pipeline([
    ('preprocessing', preprocessing_pipeline),
    ('kmeans', KMeans(n_clusters=optimal_clusters, random_state=42))
])

# Clustering pipeline dengan DBSCAN
clustering_pipeline_dbscan = Pipeline([
    ('preprocessing', preprocessing_pipeline),
    ('dbscan', DBSCAN(eps=0.5, min_samples=5))
])

# Clustering pipeline dengan Hierarchical Clustering
clustering_pipeline_hierarchical = Pipeline([
    ('preprocessing', preprocessing_pipeline),
    ('hierarchical', AgglomerativeClustering(n_clusters=optimal_clusters))
])

# Select only numerical features for clustering
numerical_features = order_cleaned.select_dtypes(include=['number'])

# Fit pipelines using only numerical features
order_cleaned['Cluster_KMeans_Pipeline'] = clustering_pipeline_kmeans.fit_predict(numerical_features)
order_cleaned['Cluster_DBSCAN_Pipeline'] = clustering_pipeline_dbscan.fit_predict(numerical_features)
order_cleaned['Cluster_Hierarchical_Pipeline'] = clustering_pipeline_hierarchical.fit_predict(numerical_features)
```

```python
# Evaluasi dengan silhouette score
silhouette_kmeans = silhouette_score(scaled_features, order_cleaned['Cluster_KMeans'])
silhouette_dbscan = silhouette_score(scaled_features, order_cleaned['Cluster_DBSCAN'])
silhouette_hierarchical = silhouette_score(scaled_features, order_cleaned['Cluster_Hierarchical'])

# Evaluasi silhouette score dari pipeline
silhouette_kmeans_pipeline = silhouette_score(scaled_features, order_cleaned['Cluster_KMeans_Pipeline'])
silhouette_dbscan_pipeline = silhouette_score(scaled_features, order_cleaned['Cluster_DBSCAN_Pipeline'])
silhouette_hierarchical_pipeline = silhouette_score(scaled_features, order_cleaned['Cluster_Hierarchical_Pipeline'])

print(f'Silhouette Score KMeans: {silhouette_kmeans}')
print(f'Silhouette Score DBSCAN: {silhouette_dbscan}')
print(f'Silhouette Score Hierarchical: {silhouette_hierarchical}')
print(f'Silhouette Score KMeans (Pipeline): {silhouette_kmeans_pipeline}')
print(f'Silhouette Score DBSCAN (Pipeline): {silhouette_dbscan_pipeline}')
print(f'Silhouette Score Hierarchical (Pipeline): {silhouette_hierarchical_pipeline}')
```

```python
# KMeans Pipeline
fig_kmeans_pipeline = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_KMeans_Pipeline', title='KMeans Clustering (Pipeline)', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig_kmeans_pipeline.show()

# DBSCAN Pipeline
fig_dbscan_pipeline = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_DBSCAN_Pipeline', title='DBSCAN Clustering (Pipeline)', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig_dbscan_pipeline.show()

# Hierarchical Pipeline
fig_hierarchical_pipeline = px.scatter(order_cleaned, x='HARGA PAKET PERBULAN', y='LAMA LANGGANAN (BULAN)', color='Cluster_Hierarchical_Pipeline', title='Hierarchical Clustering (Pipeline)', labels={'HARGA PAKET PERBULAN': 'HARGA PAKET PERBULAN', 'LAMA LANGGANAN (BULAN)': 'LAMA LANGGANAN (BULAN)'})
fig_hierarchical_pipeline.show()
```
